Parte 1: Seleção das melhores variáveis para todos os modelos.

```python
pivot_df.index = pd.to_datetime(pivot_df.index)
```

```python
pivot_df = pivot_df[pivot_df.index >= '2005-01-01']
```

```python
corr = pivot_df.corr()
```

```python
corr.style.background_gradient(cmap='coolwarm', axis=None)
```

```python
stock = ['VALE3.SA',
    'PETR4.SA',
    'ITUB4.SA',
    'BBDC4.SA',
    'ABEV3.SA',
    'WEGE3.SA', 
    'B3SA3.SA', 
    'ITSA4.SA', 
    'CSAN3.SA', 
    'BRFS',
    'AAPL',
    'NVDA',
    'MSFT',
    'AMZN',
    'GOOGL',
    'META', 
    'TSLA', 
    'UNH', 
    'DHR', 
    'SPGI',]
```

```python
resultado = []

for ativo in stock:
    correlacoes = corr['close'][ativo].drop(ativo, errors="ignore")

    positivas = correlacoes[correlacoes > 0.70].sort_values(ascending=False)
    negativas = correlacoes[correlacoes < -0.45].sort_values()
    
    selecionados_pos = [(idx, float(val)) for idx, val in positivas.items()]
    selecionados_neg = [(idx, float(val)) for idx, val in negativas.items()]
    
    selecionados = selecionados_pos + selecionados_neg

    if len(selecionados) == 0:
        top4 = correlacoes.abs().sort_values(ascending=False).head(4)
        selecionados = [(idx, float(correlacoes[idx])) for idx in top4.index]

    if len(selecionados) > 12:
        if selecionados_pos and selecionados_neg:
            total = len(selecionados_pos) + len(selecionados_neg)
            quota_pos = max(1, round(12 * len(selecionados_pos) / total))
            quota_neg = 12 - quota_pos
            selecionados = selecionados_pos[:quota_pos] + selecionados_neg[:quota_neg]
        else:
            top12 = correlacoes.abs().sort_values(ascending=False).head(12)
            selecionados = [(idx, float(correlacoes[idx])) for idx in top12.index]

    max_corr_entre_selecionados = 0.75
    final_selecionados = []

    for nome, val in sorted(selecionados, key=lambda x: abs(x[1]), reverse=True):
        redundante = False
        for f_nome, _ in final_selecionados:
            if abs(corr['close'].get(nome, {}).get(f_nome, 0)) > max_corr_entre_selecionados:
                redundante = True
                break
        if not redundante:
            final_selecionados.append((nome, val))

    candidatos = correlacoes.drop([n for n, _ in final_selecionados] + [ativo], errors='ignore')
    while len(final_selecionados) < 4 and not candidatos.empty:
        candidato_nome = candidatos.abs().idxmax()
        candidato_val = float(correlacoes[candidato_nome])
        
        if all(abs(corr['close'].get(candidato_nome, {}).get(f_nome, 0)) <= max_corr_entre_selecionados
               for f_nome, _ in final_selecionados):
            final_selecionados.append((candidato_nome, candidato_val))
        
        candidatos = candidatos.drop(candidato_nome)

    resultado.append({
        "ativo": ativo,
        "selecionados": final_selecionados
    })

for r in resultado:
    print("="*50)
    print(f"Ativo: {r['ativo']}")
    print("Selecionados:")

    ordenados = sorted(r["selecionados"], key=lambda x: abs(x[1]), reverse=True)

    for i, item in enumerate(ordenados):
        print(f"{i}: {item} (type: {type(item)})")
```

```python
disponibilidade = {}

for r in resultado:
    ativo = r['ativo']
    selecionados = [nome for nome, val in r['selecionados']]
    todos_ativos = [ativo] + selecionados

    colunas_existentes = [col for col in pivot_df.columns if col[1] in todos_ativos]

    df_sub = pivot_df[colunas_existentes]
    
    datas_disponiveis = df_sub.dropna().shape[0]
    
    disponibilidade[ativo] = {
        'ativos_selecionados': selecionados,
        'datas_disponiveis': datas_disponiveis
    }

for ativo, info in disponibilidade.items():
    print("="*50)
    print(f"Ativo: {ativo}")
    print(f"Selecionados: {info['ativos_selecionados']}")
    print(f"Datas disponíveis com todos ativos presentes: {info['datas_disponiveis']}")
```
